13) Edad promedio de los alumnos (hombres y mujeres) con absentismo ('Y') cuya nota es mayor o igual a 3


```python
df_students = pd.read_csv("resources/data_studentlist.csv", header="infer")

# 1st task
male_students = df_students[df_students["Gender"] == "M"]   # Filter male students
mean_male_students = np.round(np.mean(df_students["Height"]),2)  # Calculate mean height
print(f"Mean height in male students is: {mean_male_students} cm")

# 2nd Task
female_students = df_students[df_students["Gender"] == "F"]
mean_female_students = np.round(np.mean(female_students["Height"]),2)
print(f"Mean height in female studnets is: {mean_female_students} cm")

# 3rd Task
weight_male_studens = np.round(np.mean(male_students["Weight"]),2)
print(f"Mean weight in male students is: {weight_male_studens} kg")

# 4th Task
weight_female_studens = np.round(np.mean(female_students["Weight"]),2)
print(f"Mean weight in female students is: {weight_female_studens} kg")

# 5th Task
max_male_height = np.max(male_students["Height"])
print(f"Max height in male students is: {max_male_height} cm")

# 6th Task
min_female_height = np.min(female_students["Height"])
print(f"Min height in female students is: {min_female_height} cm")

# 7th Task
weight_175_male_students = male_students[male_students["Height"] > 175]
ligth_weight_175_male_students = np.min(weight_175_male_students["Weight"])
print(f"Light weight in greater 175 cm male students is: {ligth_weight_175_male_students} kg")

# 8th Task
weight_160_female_students = female_students[female_students["Height"] < 160]
heavy_weight_160_female_students = np.max(weight_160_female_students["Weight"])
print(f"Heavy weight in less 160 cm female students is: {heavy_weight_160_female_students} kg")

# 9th Task
assistance_students = df_students[df_students["Absence"] == "N"]
mean_assistance_students = np.round(np.mean(assistance_students["Grade"]),2)
print(f"Mean grade in assistance students is: {mean_assistance_students}")

# 10th Task
absent_students = df_students[df_students["Absence"] == "Y"]
mean_absent_students = np.round(np.mean(absent_students["Grade"]),2)
print(f"Mean grade in absence students is: {mean_absent_students}")

# 11th Task
height_a_students = df_students[df_students["Bloodtype"] == "A"]
height_ab_students = df_students[df_students["Bloodtype"] == "AB"]
height_students = pd.concat([height_a_students, height_ab_students])
mean_height_students = np.round(np.mean(height_students["Height"]),2)
print(f"Mean height in students with blood type A or AB is: {mean_height_students} cm")

# 12th Task
height_male_a_students = male_students[male_students["Bloodtype"] == "A"]
height_male_ab_students = male_students[male_students["Bloodtype"] == "AB"]
height_male_students = pd.concat([height_male_a_students, height_male_ab_students])
mean_height_male_students = np.round(np.mean(height_male_students["Height"]),2)
print(f"Mean height in male students with blood type A or AB is: {mean_height_male_students} cm")

# 13th Task
grade_students = df_students[df_students["Grade"] >= 3]
years_old_students = pd.concat([absent_students, grade_students])
mean_years_old_students = np.round(np.mean(years_old_students["Age"]),2)
print(f"Mean age in students with grade >= 3 and absence is Y is: {mean_years_old_students}")
```

## Data Sales

**Desarrolla las siguientes actividades:**

1) Añade una nueva variable Amount = UnitPrice x Units y muestra las 5 primeras observaciones.
2) Calcula el precio unitario promedio de cada región. Utiliza el método `groupby()`.
3) Calcula el precio unitario promedio de cada región. Utiliza el método `pivot_table()`.
4) Calcula el precio unitario promedio para cada región en una línea de código. Utiliza el método `groupby()`.
5) Calcula el precio unitario promedio y las unidades para cada región en una línea de código. Utiliza el método
`pivot_table()`.
6) Calcula el total de unidades para cada región y tipo de artículo en una línea de código. Utiliza el método
`pivot_table()` y rellena los valores faltantes con 0.
7) Calcula el importa total de ventas para cada región y el tipo de artículo en una línea de código. Utiliza el método
`pivot_table()` y rellena los valores faltantes con 0.

```python
df_sales = pd.read_csv('resources/data_sales.csv', header='infer')


# 1st Task
df_sales["Amount"] = df_sales["Unit Price"] * df_sales["Units"]
print(df_sales["Amount"].head(5))

# 2nd Task
unit_price_by_region = np.round(df_sales.groupby("Region")["Unit Price"].mean(),2)
print(unit_price_by_region)

# 3rd Task
unit_price_by_region_pivot = np.round(df_sales.pivot_table(index="Region", values="Unit Price", aggfunc=np.mean),2)
print(unit_price_by_region_pivot)

# 4th Task
unit_price_by_region_inline = np.round(df_sales.groupby("Region")["Unit Price"].mean(),2)
print(unit_price_by_region_inline)

# 5th Task
unit_price_by_region_pivot_inline = np.round(df_sales.pivot_table(index="Region", values="Unit Price", aggfunc=np.mean),2)
print(unit_price_by_region_pivot_inline)

# 6th Task
total_units_by_region = pd.concat([df_sales.pivot_table(index="Region", values="Units", aggfunc=np.sum), df_sales.pivot_table(index="Item", values="Units", aggfunc=np.sum, fill_value=0)])
print(total_units_by_region)

# 7th Task
total_sales_by_region = pd.concat([df_sales.pivot_table(index="Region", values="Amount", aggfunc=np.sum), df_sales.pivot_table(index="Item", values="Amount", aggfunc=np.sum, fill_value=0)])
print(total_sales_by_region)
```

## Data Coffeshop

**Desarrolla las siguientes actividades:**

1) Genera una tabla de frecuencias para la variable 'yearOfStart' y visualízala por año.
    - Ordena por año.
    - Dibuja una gráfica de línea de 1997 a 2014. <= Pista: plt.xlim()
2) Divide los datos por el estado actual del negocio ('In' o 'Out'). Después, visualiza la tendencia annual de las
frecuencias de la variable 'yearOfStart'.
    - Ordena por año.
    - Dibuja dos gráficas de líneas superpuestas de 1997 a 2014.
    - Usa el objeto 'figure'.
3) A partir de los resultados de 1) y 2), ¿cómo describirías la tendencia en general?

```python
from numpy import concatenate


df_coffe = pd.read_csv('resources/data_coffeeshop.csv', header='infer', na_values=[' '])
print(df_coffe.shape)
df_coffe.head(5)

# 1st Task
absolute_frecuency = df_coffe["Year of Start"].value_counts().sort_index()

relative_frecuency = absolute_frecuency / absolute_frecuency.sum()

accumulative_frecuency = absolute_frecuency.cumsum()

accumulative_relative_frecuency = relative_frecuency.cumsum()

df_frecuency = pd.DataFrame({
    "Absolute Frecuency": absolute_frecuency,
    "Relative Frecuency": relative_frecuency,
    "Accumulative Frecuency": accumulative_frecuency,
    "Accumulative Relative Frecuency": accumulative_relative_frecuency
})

# 1st Subtask
ordered_df_frecuency = df_frecuency.sort_values(by="Absolute Frecuency", ascending=False)
print(ordered_df_frecuency)

# 2nd Subtask
plt.plot(df_frecuency["Absolute Frecuency"], color="green")
plt.xlim(1997, 2014)


# 2nd Task
current_state_in = df_coffe[df_coffe["Current State"] == "In"]
current_state_out = df_coffe[df_coffe["Current State"] == "Out"]
current_state = pd.concat([current_state_in, current_state_out])
frecuency_current_state = current_state["Year of Start"].value_counts().sort_index()

ordered_frecuency_current_state = frecuency_current_state.sort_values(ascending=True)
ordered_frecuency_current_state

plt.plot(ordered_frecuency_current_state, color="red")
plt.xlim(1997, 2014)
plt.show()
```
