Fancy indexing:

```python
arr = np.arange(100)            # Crea un array con valores de 0 a 99
arr
```

```python
arrMask = ( (arr % 5) == 0 )    # Crea una máscara booleana que es True para los múltiplos de 5
arr[arrMask]                    # Selecciona los elementos del array que son múltiplos de 5
```

```python
arrMask = ( ( (arr % 5) == 0 )  & ( arr >0) )
arr[arrMask]
```

#### 1.5. Modifying NumPy arrays:

Appending elements, rows and columns to arrays:

```python
a = np.array([1, 2, 3])
b = np.append(a, [4, 5, 6])
b
```

```python
a = np.array([[1, 2], [3, 4]])
b = np.append(a, [[9, 9]], axis=0) #axis=0 agrega en filas
b
```

```python
c = np.append(a, [[9], [9]] , axis=1) #axis=1 agrega en columnas.
c
```

Deleting array elements, rows and columns:

```python
a = np.array([[1, 2, 3],[4, 5, 6]])
np.delete(a, 0)
```

```python
np.delete(a, (0, 2, 4))
```

```python
np.delete(a, 0, axis = 0)
```

```python
np.delete(a, 1, axis = 1)
```
