```python
df = pd.read_csv('data_studentlist.csv', header='infer')
df
```

```python
# Replace the columns (header).
df.columns = ['NAME', 'GENDER' , 'AGE', 'GRADE', 'ABSENCE', 'BLOODTYPE', 'HEIGHT', 'WEIGHT']
df.head(3)
```

#### 1.1. Adding and removing columns:

```python
df['BMI'] = 10000*df['WEIGHT']/df['HEIGHT']**2   #BMI ES EL INDICE DE MASA CORPORAL
```

```python
df.head(5)
```

```python
# drop() method removes a column. But it is just a 'view'.
df.drop('BMI',axis=1)
```

```python
# The 'drop' was not permanent,
df.head(5)
```

```python
# inplace=True makes the removal permanent.
df.drop('BMI',axis=1,inplace=True)
df.head(5)
```

#### 1.2. Merging DataFrames:

```python
df_left = df.loc[:,['NAME','AGE','GENDER','GRADE','ABSENCE']]
df_left_small = df_left.loc[:10,]
df_left_small
```

```python
df_right = df.loc[:,['NAME','BLOODTYPE','WEIGHT','HEIGHT']]
df_right_small = df_right.loc[7:,]
df_right_small
```

Inner join:

```python
pd.merge(df_left_small,df_right_small,left_on='NAME', right_on = 'NAME', how='inner')  #SOLO CUANDO LOS NOMBRES ESTEN EN LAS DOS DATAFRAMES LOS UNE Y ME CREA EL RESULTADO.
                                                                        # Solo mantiene filas que coinciden en ambos lados.
```

Left join:

```python
pd.merge(df_left_small,df_right_small,left_on='NAME', right_on = 'NAME', how='left')
                                                #Mantiene todas las filas de la izquierda, con coincidencias de la derecha; SI NO HAY rellena con NaN.
```

Right join:

```python
pd.merge(df_left_small,df_right_small,left_on='NAME', right_on = 'NAME', how='right')
                                        #how='right': Mantiene todas las filas de la derecha, con coincidencias de la izquierda; SI NO HAY rellena CON NAN.
```

Full outer join:

```python
pd.merge(df_left_small,df_right_small,left_on='NAME', right_on = 'NAME', how='outer') #Mantiene todas las filas de ambos DataFrames, llenando con NaN donde no haya coincidencias.
                                                                                      #Y LOS ORGANIZA ALFABETICAMENTE.
```
