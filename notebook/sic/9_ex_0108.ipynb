### 1. MultiIndex:

```python
import pandas as pd
import numpy as np
import os
```

#### 1.1. Creating a DataFrame with MultiIndex:

```python
my_header = ['A','B','C']
```

```python
my_index_out = ['G1']*2 + ['G2']*2 + ['G3']*2
my_index_in = ['a', 'b']*3
```

```python
my_index_zipped = list(zip(my_index_out, my_index_in))
my_index_zipped
```

```python
my_multi_index = pd.MultiIndex.from_tuples(my_index_zipped)
df = pd.DataFrame(data=np.random.randn(6,3),index=my_multi_index,columns=my_header)
```

```python
df
```

#### 1.2. Indexing and slicing DataFrames with MultiIndex:

```python
df.loc['G1']
```

```python
# Indexing followed by another indexing.
df.loc['G1'].loc['a']
```

```python
# Indexing followed by another indexing.
df.loc['G1'].loc['a','B']
```

#### 2.1. Group by method:

```python
from google.colab import files
uploaded = files.upload()
df = pd.read_csv('data_studentlist.csv', header='infer')
df
```

```python
df.head(3)
```

```python
df.groupby('gender')['height'].mean()
#df.groupby('gender'): Agrupa el DataFrame df por la columna 'gender'.
#Esto significa que el DataFrame se divide en grupos según los valores únicos de la columna gender (por ejemplo, "Masculino", "Femenino").
#['height']: Después de agrupar por gender, selecciona la columna 'height' (altura). Esto significa que solo se calculará la media de la columna 'height' para cada grupo de género.
#mean(): Calcula la media (promedio) de la columna 'height' para cada grupo formado por gender.
```

```python
df.groupby('gender')[['height','weight']].mean()
```

```python
df.groupby('gender')[['grade','age']].std()
```

```python
df.groupby('gender')['height'].describe()
```

```python
sr = df.groupby(['gender','bloodtype'])['height'].mean()     # MultiIndexed result.
sr
```
